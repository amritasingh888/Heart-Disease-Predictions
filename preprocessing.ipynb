```python
q1=np.quantile(data["trestbps"], 0.25)
q3=np.quantile(data["trestbps"], 0.75)  
```

```python
import numpy as np
def roy(data, x):
    q1 = np.quantile(data[x], 0.25)
    q3 = np.quantile(data[x], 0.75)
    iqr = q3 - q1
    
    lowest_value = q1 - 1.5 * iqr
    highest_value = q3 + 1.5 * iqr
    
    filtered_data = data[(data[x] > lowest_value) & (data[x] < highest_value)]
    
    print("Removing outliers completed...")
    sns.boxplot(data[x])
    return filtered_data
```

```python
data=roy(data,['chol'])
```

```python
q1=np.quantile(data["thalach"], 0.25)
q3=np.quantile(data["thalach"], 0.75)  
```

```python
iqr=q3-q1
```

```python
upper_value=q3+1.5*iqr
lower_value=q1-1.5*iqr
data=data[(data["thalach"]<upper_value) & (data["thalach"]>lower_value)]  
```
